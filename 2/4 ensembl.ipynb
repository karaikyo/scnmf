```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scanpy as sc
import anndata
from biomart import BiomartServer
```

```python
moudata = anndata.read_h5ad('3 clustered_mou_0829_14.h5ad')
humdata = anndata.read_h5ad('3 clustered_hum_0829_14.h5ad')
```

```python
server = BiomartServer( "www.ensembl.org/biomart" )
```

```python
server.show_databases()
```

```python
server.show_datasets()
```

```python
mou_esb = server.datasets['mmusculus_gene_ensembl'] #Mouse genes (GRCm39),
hum_esb = server.datasets['hsapiens_gene_ensembl'] #Human genes (GRCh38.p13)
```

```python
len(moudata.var_names.to_list())
```

```python
f = open("mou_res.txt", "a")

for i in range(43):
    print(i)
    response = mou_esb.search({
      'filters': {'mgi_symbol' : moudata.var_names.to_list()[i*400:i*400 + 400]},
      'attributes': ['mgi_symbol', 'ensembl_gene_id']})
    f.write(response.text)

f.close()
```
